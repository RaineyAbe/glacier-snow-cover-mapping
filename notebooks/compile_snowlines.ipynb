# Compile snowlines

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import glob
```

```python
site_name = 'RGI60-02.05157'
site_path = '/Volumes/LaCie/raineyaberle/Research/PhD/snow_cover_mapping/study-sites/' + site_name
# path to snowlines
snowlines_path = os.path.join(site_path, 'imagery', 'snowlines')
snowline_fns = sorted(glob.glob(os.path.join(snowlines_path, '*.csv')))
print('Number of snowlines = ', len(snowline_fns))
[os.path.basename(x) for x in snowline_fns]
```

```python
snowlines = pd.DataFrame()
for snowline_fn in snowline_fns:
    snowline = pd.read_csv(snowline_fn)
    # dt = os.path.basename(snowline_fn)[0:15]
    # if 'Landsat' in os.path.basename(snowline_fn):
    #     dataset = 'Landsat'
    # elif 'Sentinel-2_SR' in os.path.basename(snowline_fn):
    #     dataset = 'Sentinel-2_SR'
    # elif 'Sentinel-2_TOA' in os.path.basename(snowline_fn):
    #     dataset = 'Sentinel-2_TOA'
    # snowline['datetime'] = dt
    # snowline['dataset'] = dataset
    snowlines = pd.concat([snowlines, snowline])

snowlines['site_name'] = site_name
snowlines_fn = os.path.join(site_path, site_name + '_snowlines.csv')
snowlines.to_csv(snowlines_fn, index=False)
snowlines
```

```python
snowlines = pd.read_csv(snowlines_fn)
snowlines['datetime'] = pd.to_datetime(snowlines['datetime'], format='mixed')
plt.plot(snowlines['datetime'], snowlines['AAR'], '.')
plt.show()
```
